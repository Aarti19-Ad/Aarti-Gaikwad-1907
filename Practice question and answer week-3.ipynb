```python
# 1.List manipulation basics
```

```python
my_list = [10, 20, 30, 40, 50, 60]

# Change index 2 to 35
my_list[2] = 35

# Add 70 at the end
my_list.append(70)

# Insert 5 at the beginning 
my_list.insert(0, 5)

# Remove 40
my_list.remove(40)

# Print final list
print(my_list)
```

```python
# 2. Sum of even numbers
```

```python
numbers = [12, 7, 23, 48, 15, 66, 9]

sum_even = 0

for num in numbers:
    if num % 2 == 0:
        sum_even += num

print("Sum of even numbers:",sum_even)
```

```python
# 3. Manual list reversal
```

```python
data = ['apple', 'banana', 'cherry', 'date', 'elderberry']

reverse_list = []

for i in range(len(data)-1, -1, -1):
    reverse_list.append(data[i])

print(reverse_list)
```

```python
# 4. Finding the maximum element
```

```python
values = [8, 2, 10, 5, 17, 3, 14]

largest = values[0]

for num in values:
    if num > largest:
        largest = num

print("Largest number:",largest)
```

```python
# 5. Count occurrences
```

```python
items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']

count = 0

for i in items:
    if i == 'red':
        count += 1

print("Count of red:", count)
```

```python
# 6. Remove all occurrrences of an element
```

```python
mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5]

new_list = []

for i in mixed_numbers:
    if i != 5:
        new_list.append(i)

print(new_list)
```

```python
# 7. List palindrome check
```

```python
# List1
list1 = [1, 2, 3, 2, 1]

reverse_list = []

for i in range(len(list1)-1, -1, -1):
    reverse_list.append(list1[i])
if list1 == reverse_list:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# List2
list2 = ['a', 'b', 'c', 'b', 'd']

reverse_list = []

for i in range(len(list2)-1, -1, -1):
    reverse_list.append(list2[i])

if list2 == reverse_list:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
# 8. Merge and sort two lists (Manual sort) 
```

```python
list_a = [3, 1, 4]
list_b = [2, 5, 0]

merged = list_a + list_b

# Bubble sort
for i in range(len(merged)):
    for j in range(len(merged) -1 -i):
        if merged[j] > merged[j + 1]:
            merged[j], merged[j + 1] = merged[j + 1], merged[j]

print("Sorted list:",merged)
```

```python
# 9. Flatten a nested list
```

```python
nested_list = [1, [2, 3], 4, [5, 6, 7], 8]

flat_list = []

for i in nested_list:
    if type(i) == list:
        for num in i:
            flat_list.append(num)
    else:
        flat_list.append(i)

print(flat_list)
```

```python
# 10. Find common list
```

```python
list_p = [1, 2, 3, 4, 5]
list_q = [4, 5, 6, 7, 8]

common = []

for i in list_p:
    for j in list_q:
        if i == j:
            common.append(i)

print("Common elements:",common)
```
