```python
"""

Create a class called student
create 2-3 objects having attributes name, age

create a method which will add a skill (eg.. python, ML, java) to the student
where u can view the info of it later (i.e. skills learned by a student)

""";
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.skills = []

    def add_skill(self, skill):
        self.skills.append(skill)

    def display(self):
        print("Name :", self.name)
        print("Age :", self.age)
        print("Skills :", self.skills)
        print()


# Objects
s1 = Student("Aarti", 20)
s2 = Student("Rahul", 21)
s3 = Student("Priya", 19)

# Add skills
s1.add_skill("Python")
s1.add_skill("Java")

s2.add_skill("Machine Learning")

s3.add_skill("Python")
s3.add_skill("C++")

# Display information
s1.display()
s2.display()
s3.display()
```

Inheritance

```python
class Person:
    insti = "MITU"

    def __init__(self, age, ht):
        self.age = age
        self.ht = ht

    def show(self):
        print("Age:",self.age)
        print("Height",self.ht)
        print("Institute",self.insti)
```

```python
class Ravi(Person):

    def __init__(self, age, ht, sal):
        super().__init__(age, ht)
        self.sal = sal

    def show_sal(self):
        print(f"Salary is {self.sal}")
```

```python
a = Ravi(25, 5.9, 57000)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
class Menuapp:

    def __init__ (self):
        self.itemlist = []

    def show_menu(self):
        
        print("Menu\n 1.Add items\n2.View items\n3.Exit")

    def add_items(self):
          item = input("Enter the item name:")
          self.itemlist.append(item)

    def view_items(self):
         return self.itemlist

    
    def run(self):
        
        while True:
            
            self.show_menu()
            ch = int(input("Enter the choice:"))
            
            if ch == 1:
                self.add_items()
            elif ch == 2:
                self.view_items()
            elif ch == 3:
                break
            else:
                return 'Invalid choice'

         
```

```python
atharv = Menuapp()
```

```python
atharv.run()
```

```python
class Person:

    def __init__(self, age, name):

        self.__age = age
        self.__name = name

    #Getter

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    #Setter
    def set_age(self, age):
        self.__age = age

    def __str__(self):
       return f"Age -> {self.__age}, Name -> {self.__name}"
```

```python
# Create a menu driven app for banking simulation
# check balance, deposit money, withdraw money (you can add more things as you see fit to make the app better)
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def check_balance(self):
        print("Current Balance =", self.balance)

    def deposit(self, amount):
        self.balance += amount
        print("Deposit Successful!")
        print("New Balance =", self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawal Successful!")
            print("Remaining Balance =", self.balance)
        else:
            print("Insufficient Balance!")

# Create object
account = BankAccount(1000)

while True:
    print("\n--- Banking Menu ---")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        account.check_balance()

    elif choice == 2:
        amount = float(input("Enter amount to deposit: "))
        account.deposit(amount)

    elif choice == 3:
        amount = float(input("Enter amount to withdraw: "))
        account.withdraw(amount)

    elif choice == 4:
        print("Thank You for using Banking App!")
        break

    else:
        print("Invalid Choice!")
```
