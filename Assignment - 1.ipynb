```python
# 5. Take a year as input and print whether it is a leap year or not.
```

```python
year = int(input("Enter the year:"))
if(year % 4 == 0) or (year % 100 != 0) and (year % 400 == 0):
    print("Leap year")
else:
    print("Not a leap year")
```

```python
# 6. Build a simple calculator - take two numbers and an operator (+, -, *, /) as input, perform
# the operation, and print the result. Print an error message for invalid operators or 
# division by zero.
```

```python
a = int(input("Enter the first number:"))
b = int(input("Enter the second number:"))
op = input("Enter operator(+, -, *, /):")
if op == "+":
    print("Result:",a + b)
elif op == "-":
    print("Result:",a - b)
elif op == "*":
    print("Result:",a * b)
elif op == "/":
    if b != 0:
           print("Result:",a / b)
    else:
        print("Division by zero is not possible..!!")
else:
    print("Invalid operator")
 
    
```

```python
# 8. Take a positive integer n as input and print all prime numbers fron 2 to n.
```

```python
n = int(input("Enter the number:"))
for num in range(2, n + 1):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```

```python
# 9. Keep accepting numbers from the users until the enter 0. Then print total count of umbers 
# entered, their sum, and their average. Do not include 0 in the calculations.
```

```python
count = 0
total = 0
while True:
    num = int(input("Enter number (0 to stop):"))
    if num == 0:
        break
    count += 1
    total += num
if count > 0:
    avg = total / count
    print("Count:",count)
    print("Sum:",total)
    print("Avearge:",avg)
else:
    print("No numbers entered..!!")
```

```python
# 10. Take n as input and print a triangle pattern of n rows made of asterisks, where row i has
# (2i-1) stars centered appropriately.
```

```python
n = int(input("Enter number of rows:"))
for i in range(1, n +1):
    print(" " * (n - i) + "*" * (2 * i -1))
```
