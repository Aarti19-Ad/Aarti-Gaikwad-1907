```python
"""

Read the time in seconds as user input
and print the time in the format as hours:minutes:seconds

eg...
input -> 125 seconds
output -> 00:02:05

""";
```

```python
t = int(input("Enter the seconds:"))
hours = t // 3600
remaining = t % 3600
minutes = remaining // 60
seconds = remaining % 60
print(f"{hours:02d}:{minutes:02d}:{seconds:02d}")
```
