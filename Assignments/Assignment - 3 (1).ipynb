```python
# 1. Bank account class
```

```python
class BankAccount:
    
    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance..!!")
        else:
            self.balance -= amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Holder: {self.account_holder}, Acc No: {self.account_number}, Balance:{self.balance}"

a1 = BankAccount("Aarti", 101, 5000)
a2 = BankAccount("Sanskruti", 102, 3000)

a1.deposit(2000)
a1.withdraw(1000)

a2.withdraw(4000)
a2.deposit(500)

print(a1)
print(a2)
```

```python
# 2. Student class
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        per = self.calculate_percentage()
        if per >= 80:
            return "A"
        elif per >= 60:
            return "B"
        elif per >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("Name:",self.name)
        print("Roll No:",self.roll_no)
        print("Percentage:",self.calculate_percentage())
        print("Grade:",self.get_grade())
        print()

s1 = Student("Aarti", 1, [80, 85, 90, 75, 70])
s2 = Student("Sanskruti", 2, [65, 60, 70, 68, 63])
s3 = Student("priya", 3, [35, 40, 30, 45, 38])

Student = [s1, s2, s3]

for s in Student:
    s.display()
```

Section B - Encapsulation

```python
# 3. Employee class (encapsulation)
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.set_salary(salary)

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_emp_id(self):
        return self.__emp_id

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary must be >= 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100


e = Employee("Aarti", 101, 20000)

print(e.get_salary())

e.give_raise(10)

print(e.get_salary())
```

```python
# 4. Secure vault class
```

```python
class SecureVault:
    
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")

vault = SecureVault(1234)

vault.add_entry(1234,"Email","aarti@gmail.com")

print(vault.get_entry(1234,"Email"))

vault.add_entry(1111,"Phone","9579064595")

print(vault.get_entry(1111,"Email"))
```

Section C - Inheritance and method overriding

```python
# 5. Inheritance and method overriding (shape)
```

```python
import math

class Shape:

    def area(self):
        return 0

    def describe(self):
        print("I am a shape")

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi*self.radius**2

    def describe(self):
        print("I am a circle")

class Rectangle(Shape):
    
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length*self.width

    def describe(self):
        print("I am a rectangle")

class Triangle(Shape):
    
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5*self.base*self.height

    def describe(self):
        print("I am a triangle")

c = Circle(5)
r = Rectangle(4, 6)
t = Triangle(3, 8)

c.describe()
print("Area:",c.area())

r.describe()
print("Area:",r.area())

t.describe()
print("Area:",t.area())
```

```python
# 6. Vehicle, car and electric car
```

```python
class Vehicle:

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")

class Car(Vehicle):

    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(f"{self.make} {self.model} engine started")

class ElectricCar(Vehicle):

    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self): 
        print(f"Silent start - battery at {self.battery_capacity} kWh")

c = Car("Toyota", "Camry", 2023, 4)
e = ElectricCar("Tesla", "Model3", 2024, 75)

c.start_engine()
e.start_engine()
```

Section D - Polymorphism

```python
class Animal:

    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

class Cow(Animal):
    def sound(self):
        print("Mooo")

def make_all_sounds(animals):
    for animal in animals:
        animal.sound()

animals = [Dog(), Cat(), Cow()]
make_all_sounds(animals)
```

```python
# 8. Calculator and operator overloading
```

```python
class Calculator:

    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c

calc = Calculator()

print(calc.add(2, 3))
print(calc.add(2, 3, 4))

print(calc.multiply(2, 3))
print(calc.multiply(2, 3, 4))

class Vector:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2
print(v3)
```
