Section A - Exception handling

```python
# 1. safe_divide(a, b)
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print("Result:", result)
        
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
        
    except TypeError:
          print("Error: Invalid data type")

    finally:
          print("Operation attempted")

safe_divide(10, 2)
safe_divide(10, 0)
safe_divide(10, "A")
```

```python
# 2. validate_age(age)
```

```python
class InvalidAgeError(Exception):
    pass

def validate_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError("Invalid age")
    print("Valid age:", age)

ages = [25, -5, 130]

for age in ages:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print(e)
```

```python
# 3. User input validation
```

```python
user_inputs = ['10', 'abc', '0', '5']

for i in user_inputs:
    try:
        result = 100/int(i)
        print("Result:",result)

    except ValueError:
        print("Value error:",i)

    except ZeroDivisionError:
        print("Zero division error")

    except Exception:
        print("Some other error")
```

```python
# 4. Array 1 to 20
```

```python
import numpy as np

arr = np.arange(1, 21)

a = arr.reshape(4, 5)
print(a)

print("Row sum :",a.sum(axis = 1))
print("Column sum:",a.sum(axis = 0))

print("Greater than 10 =",arr[arr > 10])

arr[arr % 2 == 0] = -1
print(arr)
```

```python
# 5. Matrix operation
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])

B = np.array([[5, 6], [7, 8]])

print("Addition:")
print(A + B)

print("Subtraction:")
print(A - B)

print("Multiplication:")
print(A * B)

print("Matrix multiplication:")
print(A @ B)

print("Transpose:")
print(A.T)

print("Determinant:")
print(np.linalg.det(A))

print("Inverse:")
print(np.linalg.inv(A))
```

```python
# 6. Random array
```

```python
import numpy as np

np.random.seed(42)

arr = np.random.randint(1, 101, 15)

print(arr)

print("Mean:",np.mean(arr))

print("Median:",np.median(arr))

print("Std:",np.std(arr))

print("Min:",np.min(arr))

print("Max:",np.max(arr))

print("Descending:",np.sort(arr)[::-1])

print("Largest 3 indices:",np.argsort(arr) [-3:])

norm = (arr - arr.min()) / (arr.max() - arr.min())
print(norm)
```

 Section C - Pandas basics

```python
# 7. Dataframe  operations
```

```python
import pandas as pd

data = {
    'Name':['Aarav', 'Priya', 'Karan', 'Meera', 'Sahil'],
    'Age':[23, 21, 25, 22, 24],
    'Department':['AI', 'IT', 'Data science', 'AI', 'Web dev'],
    'Score':[88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

# a
print(df.head(3))
print(df.dtypes)

# b 
df['Grade'] = df['Score'].apply(
    lambda x: 'A' if x >= 90 else('B' if x >= 75 else 'C')
) 

# c
print(df.sort_values(by='Score',ascending=False))

# d
print(df[df['Department'] == 'AI'])
```

```python
# 8. Same dataframe operations
```

```python
# a
print(df.groupby('Department')['Score'].mean())

# b
print(df.loc[df['Score'].idxmax()])

# c
df.rename(columns = {'Score':'Marks'},inplace=True)

# d
df.drop('Age', axis=1, inplace=True)
print(df)
```

```python
# 9. Missing values handling
```

```python
import pandas as pd
import numpy as np

data = {
    'Product':['Pen', 'Notebook', 'Eraser', 'Sharpener', 'Ruler'],
    'Price':[10, np.nan, 5, 8, np.nan],
    'Stock':[100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

# a
print(df.isnull().sum())

# b
df['Price'].fillna(df['Price'].mean(),inplace = True)

# c
df.dropna(subset=['Stock'],inplace=True)

# d
df.to_csv('Cleaned_products.csv',index=False)
print(df)
```
