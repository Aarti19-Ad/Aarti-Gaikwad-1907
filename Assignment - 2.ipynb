```python
# 1. Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with 
# word_count, char_count(excluding spaces), unique_words(case-insensitive), and most_frequent_word. Test it with:
# "The quick brown fox jumps over the lazy dog the fox".
```

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()
    
    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
        
    most_frequent_word = max(freq, key=freq.get)
    
    return{
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", " ")),
        "unique_words": len(set(words)),
        "most_frequent_word": most_frequent_word
    }

sentence = "The quick brown fox jumps over the lazy dog the fox"
print(analyze_sentence(sentence))
```

```python
# 2. Given the string below, do the following and print each result:
# text = "A man a plan a canal panama"
# a) Reverse the string without using slicing. 
# b) Check if it is a palindrome, ignoring case and spaces.
# c) Replace every vowel with *.
# d) Count how many times each vowel appears.
```

```python
text = "A man a plan a canal Panama"

# a) Reverse without slicing
rev = ""
for ch in text:
    rev = ch + rev
print("Reverse:",rev)

# b) Palindrome check
clean = text.replace(" ", "").lower()

rev_clean = ""
for ch in clean:
    rev_clean = ch + rev_clean
if clean == rev_clean:
    print("Palindrome")
else:
    print("Not palindrome")

# c) Replace vowels with *
vowels = "AEIOUaeiou"
new_text = ""

for ch in text:
    if ch in vowels:
        new_text += "*"
    else:
        new_text += ch
print("Replace vowels:",new_text)

# d) Count vowels
count = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}

for ch in text.lower():
    if ch in count:
        count[ch] += 1
print("Vowel count:",count)
```

```python
# 3. Write a function format_report(data) that takes a list of tuples(name, score, grade) and prints a
# formatted table where name is left-aligned (width 15), score is right-aligned (width 6), amd grade is
# centered (width 8). Use this data:
# data = [{'Alice', 92, 'A'}, {'Bob', 78, 'B+'}, {'Charlie, 55, 'C''}, {'Diana', 100, 'A+'}].
```

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")

    for name, score, grade in data:
         print(f"{name:<15}{score:>6}{grade:^8}")
data = [
    ('Alice', 92, 'A'),
    ('Bob', 78, 'B+'),
    ('Charlie', 55, 'C'),
    ('Diana', 100, 'A+'),
]

format_report(data)
```

```python
# 4. Write a function list_ops(nums) that:
# a) Removes duplicates while preserving the original order.
# b) Returns a new list with only elements at even indicates, squared.
# c) Flattens the nested list into a single list using list comprehension.
# nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5] nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
```

```python
nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

def list_ops(nums):
    # a) Remove duplicates
    unique = []
    for i in nums:
        if i not in unique:
            unique.append(i)
                         
    # b) Square elements at even indices
    even_sq = [nums[i]**2 for i in range(0, len(nums), 2)]

    # c) Flatten nested list
    flat = [x for sub in nested for x in sub]

    return unique, even_sq, flat

u, e, f = list_ops(nums)
print("Unique:",u)
print("Even index squared:",e)
print("Flattened:",f)
```

```python
# 7. Write a function inventory_manager(records) that takes a list of tuples (product, category, quantity, price) and returns 
# a dictionary grouped by category. Each category should have total_items, total_value, and a list of product names. Use:
# records = [('Laptop', 'Electronics', 10, 55000), ('Mouse', 'Electronics', 50, 499), ('Desk', 'Furniture', 8, 12000), 
# ('Chair', 'Furniture', 15, 4500), ('Pen', 'Stationary' 200, 2000)]
```

```python
records = [
    ('Laptop', 'Electronics', 10, 55000),
    ('Mouse', 'Electronics', 50, 499),
    ('Desk', 'Furniture', 8, 12000),
    ('Chair', 'Furniture', 15, 4500),
    ('Pen', 'Stationary', 200, 20),
]

def inventory_manager(records):
    result = {}

    for product, category, qty, price in records:
        if category not in result:
            result[category] = {
                'total_items': 0,
                'total_value': 0,
                'products': []
            }
        result[category]['total_items'] += qty
        result[category]['total_value'] += qty * price
        result[category]['products'].append(product)

    return result

print(inventory_manager(records))
```

```python
# 8. Implement the following:
# a) A function power_all(*args, exp=2) that returns a list of each number raised to exp.
# b) A function profile(**kwargs) that prints each key-value pair as key : value.
# c) Use a lambda inside sorted() to sort the list of dicts below by cgpa in descending order.
# students = [{"name": "Arjun", "cgpa": 8.4}, {"name": "Sanya", "cgpa": 9.1}, {"name": "Dev", "cgpa": 7.8}]
```

```python
# a)
def power_all(*args, exp=2):
    return [num ** exp for num in args]
result = power_all(2, 3, 4)
print(result)

#b) 
def profile(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

profile(Name="Aarti", Age=17, City="Pune")

#c
students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]

sorted_students = sorted(
    students,
    key=lambda x: x["cgpa"],
    reverse = True
)
print(sorted_students)
```

```python
# 9. Create a module stats_util.py with three functions:
# a) mean(data) - arithmetic mean.
# b) variance(data) - population variance without using the statistics module.
# c) normalize(data) - min - max normalization returning values between 0 and 1.
# In a separate main.py, import and use these functions on the dataset below. Also compute the standard 
# deviation using the math module.
# dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]
```

```python
%%writefile stats_utils7.py

def mean(data):
    return sum(data)/len(data)

def variance(data):
    m = mean(data)
    return sum((x-m)**2 for x in data)/len(data) 

def normalize(data):
    mn = min(data)
    mx = max(data)
    return [(x-mn)/(mx-mn) for x in data]
```

```python
import math 
from stats_utils7 import mean, variance, normalize

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

print("Mean:",mean(dataset))
print("Variance:",variance(dataset))
print("Normalized data:",normalize(dataset))
print("Standard Deviation:",math.sqrt(variance(dataset)))
```

```python
# 10. Complete the following:
# a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple and
# quintuple using it.
# b) Write a decorator @timer that prints the execution time of any function it wraps. Apply it to slow_sum(n) 
# which computes the sum of squares from 1 to n.
# c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative. Apply it to safe_sqrt(x). 
```

```python
# a)
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply
triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(10))
print(quintuple(10))

# b)
import time

def timer(func):
    def wrapper(n):
        start = time.time()
        result = func(n)
        end = time.time()

        print("Execution Time =",end - start)
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(i*i for i in range(1,n+1))

print(slow_sum(5))

# c)
import math

def validate_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Number must be positive")
        return func(x)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)
print(safe_sqrt(25))
print(safe_sqrt(-7))
```
