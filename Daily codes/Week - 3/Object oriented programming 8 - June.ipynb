```python
class Person:

    def __init__(self, name, age, ht):
        self.name = name
        self.__age = age #Private variable
        self.height = ht
    def show_info(self):
        return f"name -> {self.name}, age -> {self.__age}, height -> {self.height}"
```

```python
p1 = Person('Aarti', 17, 5.8)
p2 = Person('Sanskruti', 16, 4.8)
```

```python
p1.show_info()
```

```python
p2.show_info()
```

```python
"""

Create a class called Employee =>
attributes -> company, name, monthly_salary, job_designation
here all the objects of this company will have the same common company
make salary as private

create 3 objects -> e1, e2, e3 with given attributes

method -> 1.method to calculate the anual salary
          2.method to show entire data

""";
```

```python
class Employee:

    def __init__(self, company, name, monthly_salary, job_designation):
        self.company = company
        self.name = name
        self.__monthly_salary = monthly_salary
        self.job_designation = job_designation
    def show_info(self):
        return f"company -> {self.company}, name -> {self.name}, monthly_salary -> {self.__monthly_salary}, job_designation -> {self.job_designation}"
```

```python
e1 = Employee('Mitu', 'Aarti', 25000, 'Manager')
e2 = Employee('Linkcode', 'Priya', 26000, 'Manager')
e3 = Employee('Bitcode', 'Anuja', 27000, 'Manager')
```

```python
e1.show_info()
e2.show_info()
e3.show_info()
```
