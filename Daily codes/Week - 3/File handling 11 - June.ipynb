```python
# Write()
```

```python
sent = "Frodo took the ring. he started his journey to mount doom."
```

```python
f = open('myfile.txt', 'w')
f.write(sent)
f.close()
```

```python
# Write into a myfile.txt with context manager
# Use write function and add 3 separate lines
```

```python
with open('myfile.txt', 'w') as f:
    f.write('this is first line\n')
    f.write('this is second line\n')
    f.write('this is third line\n')
```

```python
# Simple file copy operation
```

```python
s = open('sample.txt', 'r')
f = open('myfile.txt', 'w')
```

```python
contents = s.read()
f.write(contents)
f.close()
```
