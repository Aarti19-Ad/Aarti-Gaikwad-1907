```python
#Read purchase cost and selling cost from the user
#State profit or loss and how much
```

```python
pc = int(input("Enter the pc:"))
sc = int(input("Enter the sc:"))
if sc > pc:
    profit = sc - pc
    print("Profit:",profit)
elif pc > sc:
    loss = pc - sc
    print("Loss:",loss)
else:
      print("No profit and loss")
```

```python
"""
Menu driven ->
Take a string from the user
Menu -> 1.upper, 2.lowercase, 3.replace a particular part of the string for third option it should which part
""";
```

```python
text = input("Enter the string:")
print("1. Uppercase")
print("2. Lowercase")
print("3. Replace")
ch = int(input("Select your choice:"))
if ch == 1:
    print(text.uppercase())
elif ch == 2:
    print(text.lower())
elif ch == 3:
    old = input("Enter word to replace:")
    new = input("Enter new word:")
    print(text.replace(old,new))
else:
    print("Invalid choce..!!")
```
