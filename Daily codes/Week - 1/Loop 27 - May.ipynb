```python
#Write a program to find factorial of a number using while loop
```

```python
num = int(input("Enter the number"))
fact = 1

while num > 0:
    fact = fact * num
    num -= 1

print(fact)
```

```python
#Factorial using for loop
```

```python
num = int(input("Enter the number"))

fact = 1

for n in range(1, num + 1):
    fact = fact * n

print(fact)
```

```python
#Break, continue
```

```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
    print("ok")
```
