```python
"""
Use a loop and if-else to check if a specific character is present in the string or not
get output for both if the character is present and if it is not present
take 2 characters to check in the string

user input -> string, character1, character2
""";
```

```python
sent = str(input("Enter the string:"))
char1 = str(input("Enter the ch1:"))
char2 = str(input("Enter the ch2:"))

found_char1 = False
found_char2 = False

for char in sent:
    if char == char1:
        found_char1 = True
    if char == char2:
        found_char2 = True

if found_char1:
    print(f"The character '{char1}' is present in the string")
else:
  print(f"The character '{char2}' is not present in the string")


if found_char2:
    print(f"The character '{char1}' is present in the string")
else:
  print(f"The character '{char2}' is not present in the string")
```

```python
#Vowel counter
```

```python
s = input("Enter a string: ")

vowels = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}
total = 0

for ch in s.lower():
    if ch in vowels:
        vowels[ch] += 1
        total += 1

print("Total vowels =", total)
print("Vowel counts =", vowels)

#Take a string from the user. 
#Check each character using a for loop.
#If it is a vowel, increase its count in the dictionary.
#Print total vowels and vowel-wise count.
```

```python
#Palindrome checker with list
```

```python
s = input("Enter a string: ")

rev_list = []

for ch in s:
    rev_list.insert(0, ch)

rev_str = "".join(rev_list)

if s == rev_str:
    print("Palindrome")
else:
    print("Not Palindrome")
#Take a string as input.
#Reverse it using a list.
#Compare original and reversed strings.
#If both are same, it is a palindrome
```

```python
#Character frequency counter
```

```python
s = input("Enter a string: ")

freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)

#Take a string from the user.
#Use a dictionary to store character counts.
#Increase count when a character repeats.
#Print the frequency of each character
```
