```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([10, 23, 29, 34, 39, 45, 50])
y = np.array([14, 21, 31, 43, 48, 55, 58])
z = np.array([56, 50, 44, 37, 26, 20, 16])
```

```python
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line chart: x vs y")
plt.plot(x, y, marker = 'o', mfc = 'red', ls= '--')
plt.grid();
# o, *, H, <, >, ^
```

```python
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line chart: x vs y")
plt.plot(x, z, marker = 'o', mfc = 'orange', ls = '--')
plt.grid();
# o, *, H, <, >, ^
```

```python
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line chart: x vs y")
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--', label = '2024')
plt.plot(x, z, marker = 'o', mfc = 'black', ls = '--', color = 'orange', label = '2025')
plt.grid();
plt.legend();
# o, *, H, <, >, ^
```

```python
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line chart: x vs y")
plt.plot(x, y, marker = 'o', mfc = 'red', ls = '--')
plt.grid();

plt.subplot(2, 2, 2)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line chart: x vs y")
plt.plot(x, z, marker = 'o', mfc = 'orange', ls = '--')
plt.grid();
```
