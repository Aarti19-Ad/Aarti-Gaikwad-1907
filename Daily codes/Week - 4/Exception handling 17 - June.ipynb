```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
except Exception:
    print("Something went wrong")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
```

```python
try:
    f = open('a1.txt')
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("Run complete")
```

```python
try:
    f = open('a1.txt')
    if f.name == 'a1.txt':
        raise Exception
except FileNotFoundError:
    print("File does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("Run complete")
```

```python
# Take 2 numbers as user input 
# Write a program for division of these 2 numbers 
# Handle Zero division error and value error
```

```python
a = int(input("Enter first number"))
b = int(input("Enter second number"))
try:
    c = a / b
    print(c)
except Exception as e:
    print(e)
```

```python
a = int(input("Enter first number"))
b = int(input("Enter second number"))
try:
    c = a/b
    print(c)
except ZeroDivisionError:
    print("Value is divide by zero")
except ValueError:
    print("Value is not find")
finally:
    print("Run complete")
```
