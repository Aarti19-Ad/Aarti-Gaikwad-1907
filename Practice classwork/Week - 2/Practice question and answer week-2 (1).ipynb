```python
#Vowel counter
```

```python
def vowel_counter(text):
    vowels = "aeiou"
    count = 0
    vowel_dict = {"a":0,"e":0,"i":0,"o":0,"u":0}

    for ch in text.lower():
        if ch in vowels:
            count += 1
            vowel_dict[ch] += 1

    print("Total Vowels =", count)
    print("Vowel Counts =", vowel_dict)

vowel_counter("Hello World")
```

```python
#Palindrome checker with list
```

```python
def palindrome_check(text):
    rev_list = []

    for ch in text:
        rev_list.insert(0, ch)

    rev_text = "".join(rev_list)

    if text == rev_text:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome_check("madam")
```

```python
#Character frequency counter
```

```python
def char_frequency(text):
    freq = {}

    for ch in text:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    print(freq)

char_frequency("banana")
```

```python
#Word length classification
```

```python
def classify_words(sentence):
    short_words = []
    medium_words = []
    long_words = []

    words = sentence.split()

    for word in words:
        if len(word) < 5:
            short_words.append(word)
        elif 5 <= len(word) <= 8:
            medium_words.append(word)
        else:
            long_words.append(word)

    print("Short:", short_words)
    print("Medium:", medium_words)
    print("Long:", long_words)

classify_words("I love Python programming language")
```

```python
#Substring with specific Start / End
```

```python
def matching_strings(lst):
    result = []

    for s in lst:
        if s.startswith("py") and s.endswith("thon"):
            result.append(s)

    print(result)

matching_strings(["python", "pythons", "pyathon", "java"])
```

```python
#Unique word extractor
```

```python
import string

def unique_words(paragraph):
    paragraph = paragraph.lower()

    for p in string.punctuation:
        paragraph = paragraph.replace(p, "")

    words = paragraph.split()
    unique = set()

    for word in words:
        unique.add(word)

    result = sorted(list(unique))
    print(result)

unique_words("Python is easy. Python is powerful!")
```

```python
#Even or odd indexed character separator
```

```python
def separate_chars(text):
    even = []
    odd = []

    for i, ch in enumerate(text):
        if i % 2 == 0:
            even.append(ch)
        else:
            odd.append(ch)

    print("Even Index Characters:", even)
    print("Odd Index Characters:", odd)

separate_chars("Python")
```

```python
#String anagram checker
```

```python
def anagram_check(str1, str2):
    list1 = sorted(list(str1.lower()))
    list2 = sorted(list(str2.lower()))

    if list1 == list2:
        print("Anagram")
    else:
        print("Not Anagram")

anagram_check("listen", "silent")
```

```python
#Missing character finder (Alphabet check)
```

```python
def missing_letters(text):
    letters = set("abcdefghijklmnopqrstuvwxyz")

    for ch in text.lower():
        if ch in letters:
            letters.remove(ch)

    print("Missing Letters:", letters)

missing_letters("hello")
```

```python
#Custom string reversal
```

```python
def reverse_words(sentence):
    words = sentence.split()

    reversed_words = words[::-1]

    result = " ".join(reversed_words)

    print(result)

reverse_words("Hello World")
```
