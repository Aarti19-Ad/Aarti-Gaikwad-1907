```python
#Set operations
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
```

```python
s1.union(s2) #Duplicates value not allowed
```

```python
#Or
```

```python
s1 | s2
```

```python
s1.intersection(s2) #Only duplicates value
```

```python
s1.difference(s2) #Only s1 under present
```

```python
s1.symmetric_difference(s2) #Both set values are allowed but same value is not allowed
```

```python
s1 = {1,2,3,4,5,6,7,8}
s2 = {4,5,6,7,8}
s3 = {1,2,3}
```
